*What is machine learning?*
- Algorithms that generate software from desired outcomes
- Training data, model form/ learning function, model/software and accuracy metric
- compose model form, fit unknown terms to data, evaluate to solve problem -> understanding model by understanding algorithms
- modern AI techniques delegate more creativity to algorithm, more interpretation to human -> use complex models with care
- limit or penalize complexity and cross validate

*Machine-learning algorithms*
- supervised learning (find mapping, e.g. classification, regression)
    - simple linear regression
    - simple logistic regression
    - Bayes' theorem
    - k-nearest neighbors
    - decision trees (series of rules, recursive approach)
- clustering (discover similiar data)
- dimensionality reduction (simple description of data)
- there is no 'algorithm to rule them all', each have their advantages and drawbacks
