## Part 1: Reading the Nucleon Form Factors

### 1.1 Physics Background and Motivations

Within the Standard Model of particle physics, nucleons (protons and neutrons), the constituents of atomic nuclei, are complex structures composed of quarks and gluons.
Their interactions are governed by quantum chromodynamics (QCD), making nucleon structure a direct and intricate result of quark-gluon dynamics.
Nucleon form factors serve as powerful tools for probing this structure.
Despite significant theoretical and experimental efforts, several unresolved issues and inconsistencies remain regarding the fundamental properties of nucleons.

Experiments studying electron scattering off protons and neutrons have a long history of providing an understanding of the structure of nucleons.
Quantitative understanding of the distribution of charge is described by the electric and magnetic form factors, $G_E$ and $G_M$.
One key issue that the physics community is looking into is the behavior of the ratio $G_E/G_M$ (a more direct product of experimental measurements, such as those from Jefferson Lab) at $Q^2 > 1\text{ GeV}^2$.
Also, where does this ratio cross zero;
is it at about $8\text{ GeV}^2$ as indicated by experiments at JLab?
What picture can give us better intuition to understand why?

The behavior of nucleon electromagnetic form factors at high momentum transfer also has implications for understanding and improving models of nucleon structure.
QCD models such as "vector-meson dominance", "chiral solitons", "Dyson-Schwinger and Faddeev equations", "relativistic constituent quarks" and more, have been employed to predict form-factor behavior but their predictions differ in the region where data are unavailable.
Methods, such nonperturbative lattice-QCD calculations, can be used to determine their magnitude and make predictions for upcoming experimental measurements.
For those who are interested in learning more on this topic, see [this reference](https://doi.org/10.1140/epja/i2015-15079-x) for a review.

Many lattice-QCD collaborations are interested in the nucleon electromagnetic form factors with respect to the nucleon radius, focusing on the small momentum transfer region of the form factors.
The electric (magnetic) charge radii of the nucleons can be obtained from the slope of the form factors at spacelike squared four-momentum transfer $Q^2=0$.
At present there is a $6\sigma$ discrepancy between the electric charge radius of the proton obtained from electronic energy levels combined with electron scattering data versus that from the Lamb shift in muonic hydrogen $E_{\mu p}(2S-2P)$.
For those who are interested in finding out more about the "proton-radius puzzle", you can watch [this video](https://www.youtube.com/watch?v=Jv7j9EtmQzE) to learn more and about how this topic is connected to atomic spectroscopy.

Another major focus of form-factor study is the nucleon axial form factors.
Understanding nucleon axial form factors is essential to predicting how nucleons respond to weak interactions, a key aspect of neutrino scattering.
These form factors are critical for interpreting data from current and future neutrino experiments like NO$\nu$A, MINER$\nu$A, and DUNE, especially at Fermilab where MINER$\nu$A has recently released new neutrino-interaction measurements.
While the axial charge, $g_A$, is precisely known from experiments on neutron beta decay, the momentum dependence of the axial form factor $G_A(Q^2)$ and the induced-pseudoscalar form factor $G_P(Q^2)$ remain less precisely determined by experimental data.
Due to the limited time, we will focus on the topic of nucleon electromagnetic form factors.
However, the analysis technique you learned in this unit can be easily extend to axial form factor as well.
The axial current form factors of the nucleon and nuclei are relevant for neutrino physics as discussed extensively in the [USQCD whitepaper on lattice QCD for neutrino physics](https://doi.org/10.1140/epja/i2019-12916-x).

## Part 1.2 Methodology

The Dirac $F_{1}$ and Pauli $F_{2}$ form factors are extracted from the matrix elements of the electromagnetic current within the nucleon state $N$ through the relation
$$\left\langle N(\vec{p}_f) | V_\mu^{\rm em} (\vec{q}) | N(\vec{p}_i)\right\rangle  =
{\overline u}_N(\vec{p}_f)\left( F_1(Q^2) \gamma_\mu
+\sigma_{\mu \nu} q_\nu
\frac{F_2(Q^2)}{2 M_N}\right)u_N(\vec{p}_i),$$
where $\vec{q}=\vec{p}_f-\vec{p}_i$ is the momentum transfer.
The discrete lattice momenta are given by $2 \pi {\bf n} / L a$ with the entries of the vector ${\bf n} \equiv (n_1, n_2, n_3)$ taking on integer values, $n_i \in \{0,L\}$.
The spacing between the momenta is controlled by the spatial lattice size, $La$. The normalization used for the nucleon spinors in Euclidean space is
$
\sum_{s} u_N(\vec p,s) \bar{u}_N(\vec{p},s) =
   \frac{E(\vec{p})\gamma_4-i\vec\gamma\cdot \vec{p} + M}{2 E(\vec{p})}$,
and the electromagnetic current is
$
V_\mu^\text{em} = \frac{2}{3} {\overline u} \gamma_\mu u - \frac{1}{3}{\overline d} \gamma_\mu d $.
In the isospin-symmetric limit, the difference of its matrix elements between a proton and a neutron state are related to the isovector form factors of the proton by the relation
$$
\left\langle p(\vec{p}_f) | \overline{u} \gamma_\mu u - \overline{d} \gamma_\mu d | p(\vec{p}_i)\right\rangle  
=
\left\langle p(\vec{p}_f) | V_\mu^\text{em} (\vec{q}) | p(\vec{p}_i)\right\rangle  
- \left\langle n(\vec{p}_f) | V_\mu^\text{em} (\vec{q}) | n(\vec{p}_i)\right\rangle .$$
The quantity we calculate on the lattice is the left-hand side of the above equation: the isovector form factors of the proton.
Throughout this class, the term isovector form factors of the proton and the $(p-n)$ form factors refer to the same quantities as defined above.
These will henceforth be analyzed in terms of the spacelike 4-momentum squared, $Q^2 = {\vec p}^2 - (E-m)^2 = -q^2$.

Experimentally, people more commonly use another set of definitions of the electromagnetic form factors, the Sachs electric $G_{E}$ and magnetic $G_{M}$ form factors that are related to the Dirac and Pauli form factors as
\begin{align}
G_E(Q^2) &= F_1(Q^2) - \frac{Q^2}{4M_N^2}F_2(Q^2); \\
G_M(Q^2) &= F_1(Q^2) + F_2(Q^2).
\end{align}
From these, the vector charge is given by
$
g_V = G_E|_{Q^2=0} = F_1|_{Q^2=0}
$
and the difference between the magnetic moment of the proton and the neutron by
$
\mu^{p} - \mu^{n} = G_M|_{Q^2=0} = (F_1 + F_2)|_{Q^2=0} = 1 + \kappa_p - \kappa_n .
$
The anomalous magnetic moments of the proton and the neutron, in units of the Bohr magneton, are known very precisely
\begin{align}
\kappa_p  &= 1.79284735(1)    & \text{(proton)}  \,, \\
\kappa_n  &= -1.91304273(45)  & \text{(neutron)} \,.
\end{align}
The electric and magnetic size of the nucleon are defined as the slope of the form
factors with respect to $Q^2$ at $Q^2=0$
$$
\langle r_{E,M}^2\rangle = -6\frac{d}{dQ^2}\left.\left(\frac{G_{E,M}(Q^2)}{G_{E,M}(0)}\right)\right|_{Q^2=0} \,.$$
The form factors $G_{E,M}$ are normalized by their values at $Q^2=0$:
$G_E(Q^2=0) \equiv g_V$ and $G_M(Q^2=0)/g_V \equiv \mu$.
This definition makes them independent of the renormalization constant $Z_V$ of the
lattice vector current, and improves the signal, because some of the systematics cancel in the ratios.
Therefore, in this class, if time allows, we will use
this equation when calculating $\langle r_E^2 \rangle$ and $\langle r_M^2 \rangle$.
Note that $Z_V g_V = 1$, because the electric charge is conserved.
A second independent estimate of $Z_V$, obtained using nonperturbative lattice calculations in the RI-sMOM scheme, where the difference between the two estimates was shown to be $\lesssim 3\%$ in the data set we will use in class.

## Part 2.1: Background Reading

To fully solve for Dirac and Pauli form factors will require more than the time allocated for this class.
For this unit, we will only look at a few special cases to quickly extract the Sachs form factors $G_{E,M}$.
Since the form factors require initial and final nucleon state to have different momentum (thus non-zero momentum transfer), we need to modify the ratio function introduced in the earlier class, ${\cal R}_{\mu}$, to the following more complicated ones
$$
{\cal R}_{\mu}(t, \tau, \mathbf{p}^\prime, \mathbf{p}) =   \frac{C^{(\text{3pt})}_{\mu}(t,\tau;\mathbf{p}^\prime,\mathbf{p})}{C^{(\text{2pt})}(\tau,\mathbf{p}^\prime)} \, \times \,
  \left[ \frac{C^{(\text{2pt})}(t,\mathbf{p}^\prime) C^{(\text{2pt})}(\tau,\mathbf{p}^\prime) C^{(\text{2pt})}(\tau-t,\mathbf{p})}{C^{(\text{2pt})}(t,\mathbf{p}) C^{(\text{2pt})}(\tau,\mathbf{p}) C^{(\text{2pt})}(\tau-t,\mathbf{p}^\prime)}
  \right]^{1/2} \,,
$$
give the desired ground-state matrix elements (ME) $\left\langle{0^\prime}|{\mathcal{O}_\Gamma}|{0}\right\rangle$ in the limits $t \to \infty$ and $(\tau-t) \to \infty$.
In the calculation of the nucleon three-point functions, we use the spin-projection operator ${\cal P}_3 = (1 + \gamma_4)(1 + i\gamma_5 \gamma_3)/2$.
With this ${\cal P}_3 $, and the vector current defined above with Euclidean
$\gamma_\mu$, the following quantities have a signal and give either the electric or the magnetic form factor:
\begin{align}
  \sqrt{2E_p(M_N+E_p)} \Re ({\cal R}_{i}) =& - \epsilon_{ij3} q_j {G}_M  \,, \\
  \sqrt{2E_p(M_N+E_p)} \Im ({\cal R}_{i}) =&   q_i {G}_E  \,, \\
  \sqrt{2E_p(M_N+E_p)} \Re ({\cal R}_{4}) =&   (M_N+E_p) {G}_E  \,.
\end{align}

Exploiting the cubic symmetry under spatial rotations, we construct two averages over equivalent 3-point correlators before doing fits to get the ground-state matrix elements: over $\Re(C_1)$ and $\Re(C_2)$ for $G_M(Q^2)$ and over $\Im(C_1)$, $\Im(C_2)$ and $\Im(C_3)$ for $G_E(Q^2)$.
We label these form factors $G_M^{V_i}$ and $G_E^{V_i}$.
Together with $G_E^{V_4}$, they constitute the three form factors analyzed.
Their extraction is straightforward as each of the three is given by a distinct three-point function.
It is important to note that the discretization artifacts and the excited-state contamination in each can be very different.
