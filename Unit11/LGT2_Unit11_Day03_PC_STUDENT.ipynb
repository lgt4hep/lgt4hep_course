# Part 1: Parton Distribution Functions and LaMET

Parton disbituion functions are important non-perturbative quantities that can be used to caluclate cross sections in high energy experiments.
Traditionally, PDFs were studied by fitting parameterized functions to the world's experimental data, in a process called "global fitting."
In the past decade, large momentum effective theory was developed and started giving us calculations of PDFs from lattice.
PDFs and LaMET are both rich and exciting on their own, and will act as a good stepping stone to understanding GPDs from LaMET in the asymmetric frame.

# Part 2: GPDs from LaMET and Asymmetric Frame GPD Caluclations

Generalized parton disbituion functions are powerful generalizations to PDFs, which provide rich 3D information about hadrons.
GPDs are fully sensitive to more specific interactions and require more complicated parameterizations than PDFs, so global analysis studies are far more limited.
LaMET generalizes easily to GPDs, so lattice is an extremely valuable tool for GPDs.
LaMET calculations of GPDs typically calculate matrix elements between hadrons states with the momentum transfer split symmetrically, but this requires solving more propagators for different momentum transfers, so a cutting edge method is to use an asymmetric split of the momentum between hadron states.
